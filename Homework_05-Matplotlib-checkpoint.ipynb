```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
import sklearn.datasets as dta


# Study data files
mouse_metadata_path = "02-Homework_05-Matplotlib_Instructions_Pymaceuticals_data_Mouse_metadata.csv"
study_results_path = "02-Homework_05-Matplotlib_Instructions_Pymaceuticals_data_Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

mouse_metadata.head()
study_results.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_regimen_group = main_df.groupby('Drug Regimen')

```

```python
drug_regimen_df = drug_regimen_group['Tumor Volume (mm3)'].mean()
drug_regimen_df
```

```python
drug_regimen_df = pd.DataFrame(drug_regimen_df)
drug_regimen_df
```

```python
drug_regimen_df = drug_regimen_df.rename(columns={'Tumor Volume (mm3)': 'Mean Tumor Volume'})
drug_regimen_df
```

```python
drug_regimen_df['Median Tumor Volume'] = drug_regimen_group['Tumor Volume (mm3)'].median()
drug_regimen_df['Tumor Volume Variance'] = drug_regimen_group['Tumor Volume (mm3)'].var()
drug_regimen_df['Tumor Volume Std. Dev.'] = drug_regimen_group['Tumor Volume (mm3)'].std()
drug_regimen_df['Tumor Volume Std. Err.'] = drug_regimen_group['Tumor Volume (mm3)'].sem()
drug_regimen_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
main_df.groupby("Drug Regimen").agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']})          
```

```python
quartile_df = quartile_df.rename(columns={'Timepoint_x': 'Timepoint', 'Timepoint_y': 'Final Timepoint'})
quartile_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)

treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
volume = quartile_df['Tumor Volume (mm3)']
```

```python
quartiles = volume.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of tumor volume is: {lowerq}")
print(f"The upper quartile of tumor volume is: {upperq}")
print(f"The interquartile range of tumor volume is: {iqr}")
print(f"The the median of tumor volumes is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
outlier_capomulin = quartile_df.loc[(quartile_df['Drug Regimen'] == 'Capomulin') | (quartile_df['Tumor Volume (mm3)'] < lower_bound) | (quartile_df['Tumor Volume (mm3)'] > upper_bound)]
outlier_capomulin
outlier_ramicane = quartile_df.loc[(quartile_df['Drug Regimen'] == 'Ramicane') | (quartile_df['Tumor Volume (mm3)'] < lower_bound) | (quartile_df['Tumor Volume (mm3)'] > upper_bound)]
outlier_ramicane
outlier_infubinol= quartile_df.loc[(quartile_df['Drug Regimen'] == 'Infubinol') | (quartile_df['Tumor Volume (mm3)'] < lower_bound) | (quartile_df['Tumor Volume (mm3)'] > upper_bound)]
outlier_infubinol
outlier_ceftamin= quartile_df.loc[(quartile_df['Drug Regimen'] == 'Ceftamin') | (quartile_df['Tumor Volume (mm3)'] < lower_bound) | (quartile_df['Tumor Volume (mm3)'] > upper_bound)]
outlier_ceftamin

print(f"Capomulin's potential outliers: {outlier_capomulin['Tumor Volume (mm3)']}")
print(f"Ramicane's potential outliers: {outlier_ramicane['Tumor Volume (mm3)']}")
print(f"Infubinol's potential outliers: {outlier_infubinol['Tumor Volume (mm3)']}")
print(f"Ceftamin's potential outliers: {outlier_ceftamin['Tumor Volume (mm3)']}")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
weight = average_df["Weight (g)"]
mean_tumor = average_df["Tumor Volume (mm3)"]
```

```python
print(f"The correlation coefficient between mouse weight and average tumor volume is {round(st.pearsonr(weight,mean_tumor)[0],2)}")
```

```python
vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(weight, mean_tumor)

vc_fit = vc_slope * weight + vc_int
```

```python
plt.scatter(weight,mean_tumor)
plt.plot(weight,vc_fit,"-",color="r")
plt.xticks(weight)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume')
plt.show()
```
